# What are the reasons for happiness?

## Introduction

### Data set description

The World Happiness Report (https://worldhappiness.report/) is a landmark survey of the state of global happiness. The first report was published in 2012, and has been published annually ever since.

The report continues to gain global recognition as governments, organizations and civil society increasingly use happiness indicators to inform their policy-making decisions. Leading experts across fields – economics, psychology, survey analysis, national statistics, health, public policy and more – describe how measurements of well-being can be used effectively to assess the progress of nations. The reports review the state of happiness in the world today and show how the new science of happiness explains personal and national variations in happiness.

## How are rankings calculated?
The rankings use data provided by the Gallup World Poll (for more information see the Gallup World Poll methodology https://www.gallup.com/178667/gallup-world-poll-work.aspx). 

The rankings are based on answers to the main life evaluation question asked in the poll. This is called the Cantril ladder: it asks respondents to think of a ladder, with the best possible life for them being a 10, and the worst possible life being a 0. They are then asked to rate their own current lives on that 0 to 10 scale.

# prep data sets

```python
import pycountry

common_features = ['Country', 'GDP', 'Health', 'Freedom', 'Generosity', 'Trust', 'Happiness']
targets = ['Low', 'Low-Mid', 'Top-Mid', 'Top']

df_all = pd.DataFrame()

def alpha3code(column):
    CODE=[]
    for country in column:
        try:
            code=pycountry.countries.get(name=country).alpha_3
           # .alpha_3 means 3-letter country code 
           # .alpha_2 means 2-letter country code
            CODE.append(code)
        except:
            CODE.append('None')
    return CODE
def add_lat_long(df_year):
    # merge country lat and long values
    location=pd.read_csv('https://raw.githubusercontent.com/melanieshi0120/COVID-19_global_time_series_panel_data/master/data/countries_latitude_longitude.csv')
    countries = location['name']
    #for country in countries
    #    if country in df_year['Country']:
    #        df_year['latitude']=location[country].latitude
    #        df_year['latitude']=location[country].latitude

def prep_frame(df_year, year):
    
    target_cols = []
    for c_col in common_features:
        for c in df_year:    
            # drop columns that are not common to the set being treated
            if (c not in common_features):
                df_year.drop(c, axis=1, inplace=True)
    
    # insert year
    df_year['Year'] = year
        
    #calculate quartiles
    df_year['Tier'] = pd.qcut(df_year['Happiness'], len(targets), labels=targets)
    df_year['Tier_num'] = pd.qcut(df_year['Happiness'], len(targets), labels=range(len(targets)))
    
    # create a column for code 
    df_year['Country_Code']=alpha3code(df_year['Country'])
    
    # insert lat and lon for country code
    
    # re-order columns
    df_year=df_year[['Country', 'Year', 'GDP', 'Health', 'Freedom', 'Generosity', 'Trust', 
             'Tier', 'Tier_num','Country_Code', 'Happiness']]
    df_year.set_index(['Country', 'Year'])
    
    return df_year
```

```python
df_all=prep_frame(data_2015,2015)
df_all=df_all.append(prep_frame(data_2016,2016), sort=False)
df_all=df_all.append(prep_frame(data_2017,2017), sort=False)
df_all=df_all.append(prep_frame(data_2018,2018), sort=False)
df_all=df_all.append(prep_frame(data_2019,2019), sort=False)
```
